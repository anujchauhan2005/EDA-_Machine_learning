```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
import numpy as np
import random

class MBGDRegressor:

    def __init__(self, learning_rate=0.01, epochs=100, batch_size=32):

        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size

    def fit(self, X_train, y_train):

        # Initialize parameters
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        for i in range(self.epochs):

            # Number of mini-batches
            for j in range(X_train.shape[0] // self.batch_size):

                # Randomly select batch indices
                idx = random.sample(range(X_train.shape[0]), self.batch_size)

                X_batch = X_train[idx]
                y_batch = y_train[idx]

                # Predictions
                y_hat = np.dot(X_batch, self.coef_) + self.intercept_

                # Compute gradients
                intercept_der = -2 * np.mean(y_batch - y_hat)

                coef_der = -2 * np.dot((y_batch - y_hat), X_batch) / X_batch.shape[0]

                # Update parameters
                self.intercept_ = self.intercept_ - self.lr * intercept_der
                self.coef_ = self.coef_ - self.lr * coef_der

        print("Intercept:", self.intercept_)
        print("Coefficients:", self.coef_)

    def predict(self, X_test):

        return np.dot(X_test, self.coef_) + self.intercept_
```
